```python
import numpy as np
import matplotlib.pyplot as plt
```

roll a dice

```python
dice = []
rolls = 1000

for i in range(rolls):
    dice.append(np.random.randint(1, 7))

count = []
for i in range(1, 7):
    count.append(dice.count(i))
    print(f'{i} - {count[i-1]}')

max_dif = max(count) - min(count)
print('max difference: ', max_dif)

plt.bar(list(range(1, 7)), count)
plt.show()
```

roll a dice twice

```python
dice = []
rolls = 1000000

for i in range(rolls):
    dice.append(np.random.randint(1, 7) + np.random.randint(1, 7))  #sum of two rolls

count = []
for i in range(2, 13):
    count.append(dice.count(i))
    print(f'{i} - {count[i-2]}')

max_dif = max(count) - min(count)
print('max difference: ', max_dif)

plt.bar(list(range(2, 13)), count)
plt.show()
```

roll a dice thrice

```python
dice = []
rolls = 1000000

for i in range(rolls):
    dice.append(np.random.randint(1, 7) + np.random.randint(1, 7) + np.random.randint(1, 7))  #sum of three rolls

count = []
for i in range(3, 19):
    count.append(dice.count(i))
    print(f'{i} - {count[i-3]}')

max_dif = max(count) - min(count)
print('max difference: ', max_dif)

plt.bar(list(range(3, 19)), count)
plt.show()
```
