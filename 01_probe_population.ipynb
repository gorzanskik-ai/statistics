```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
taxi = pd.read_parquet(path, engine='auto', columns=None, storage_options=None, use_nullable_dtypes=False)
```

```python
means = []
for i in range(1, 101):
    df_probe = taxi.sample(frac=i/100)
    probe_mean = df_probe['trip_distance'].mean()
    means.append(probe_mean)
    print(i, probe_mean)
```

```python
plt.plot(means)
plt.hlines(xmin=1, xmax=100, y=taxi['trip_distance'].mean(), linestyles='solid', colors='red')
plt.show()
```
